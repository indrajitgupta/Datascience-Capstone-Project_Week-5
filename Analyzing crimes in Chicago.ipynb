```python
chicago_crime=pd.get_dummies(df_data[['Primary Type']],prefix="",prefix_sep="")
```

```python
chicago_crime['Borough']=df_data['Block']
```

```python
fixed_col=[chicago_crime.columns[-1]]+list(chicago_crime.columns[:-1])
chicago_crime=chicago_crime[fixed_col]
```

```python
chicago_grouped=chicago_crime.groupby('Borough').mean().reset_index()
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
```

```python
num_top_crimes = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Borough']
for ind in np.arange(num_top_crimes):
    try:
        columns.append('{}{} Most Common Crime'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Crime'.format(ind+1))

# create a new dataframe
neighborhoods_crime_sorted = pd.DataFrame(columns=columns)
neighborhoods_crime_sorted['Borough'] = chicago_grouped['Borough']

for ind in np.arange(chicago_grouped.shape[0]):
    neighborhoods_crime_sorted.iloc[ind, 1:] = return_most_common_venues(chicago_grouped.iloc[ind, :], num_top_crimes)
neighborhoods_crime_sorted.set_index('Borough')
neighborhoods_crime_sorted.head(6)
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_nbr=5
chicago_cluster=chicago_grouped.drop('Borough',1)
k_means=KMeans(n_clusters=cluster_nbr,random_state=0).fit(chicago_cluster)
k_means.labels_[0:10]
```

```python
neighborhoods_crime_sorted.insert(0,'Cluster Labels',k_means.labels_)
```

```python
neighborhoods_crime_sorted.head()
```

```python
chicago=df_data[['Block','Latitude','Longitude']]
```

```python
#newyork.reset_index(inplace=True)
#neighborhoods_crime_sorted.reset_index(inplace=True)
chicago_merged=pd.merge(chicago,neighborhoods_crime_sorted,left_on='Block',right_on='Borough',how='inner')
```
